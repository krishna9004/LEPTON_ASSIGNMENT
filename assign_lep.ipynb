```python
import pandas as pd
import numpy as np
import requests
from bs4 import BeautifulSoup
from time import sleep
from random import randint
```

```python
headers={"Accept-Language": "en-US,en;q=0.5"}
```

```python
Address=[]
Timing=[]
Store_Name=[]
Phone_Number=[]
```

```python
pages=np.arange(1,67,1)
```

```python
for page in pages:
      page = requests.get("https://store-in.puma.com/?page="+str(page)+"")
      soup = BeautifulSoup(page.text,'html.parser')
      store_name_data = soup.findAll('div',attrs={'class': 'store-info-box'})
      sleep(randint(2,8))
      for store in store_name_data:
            address_store=store.ul.find('li',class_="outlet-address").find('div',class_="info-text").text
            Address.append(address_store)
            
            time=store.find('li',class_='outlet-timings').find('div',class_='info-text').text
            Timing.append(time)
            
            name=store.ul.find_all('li')[1].find('div',class_="info-text").text
            Store_Name.append(name)
            
            phone=store.find('li',class_='outlet-phone').find('div',class_="info-text").text
            Phone_Number.append(phone)
```

```python
store_name_data=pd.DataFrame({"name":Store_Name,"address":Address,"time":Timing,"phone":Phone_Number})
```
